```python
import pylab as pl
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%pylab inline
```

```python
datestring = '201501'
df = pd.read_csv(os.getenv("PUIDATA") + "/" + datestring + '-citibike-tripdata.csv')
df = df
df.head()
```

```python
datestring_1 = '201508'
df_1 = pd.read_csv(os.getenv("PUIDATA") + "/" + datestring_1 + '-citibike-tripdata.csv')
df_1.head()
```

```python
df['date'] = pd.to_datetime(df['starttime'])
df_1['date'] = pd.to_datetime(df_1['starttime'])
```

```python
df_new = df[['tripduration', 'date']]
df_new.head()
```

```python
df_1new = df_1[['tripduration', 'date']]
df_1new.head()
```

```python
df_new['day'] = df_new['date'].dt.day
df_new.head()
```

```python
winter = df_new.groupby(['day']).mean()
winter.reset_index(inplace=True)
winter.head()
```

```python
df_1new['day'] = df_1new['date'].dt.day
df_1new.head()
```

```python
summer = df_1new.groupby(['day']).mean()
summer.reset_index(inplace=True)
summer.head()
```

```python
data = summer.merge(winter, on='day')
data.columns = ['day', 'summer_tripduration', 'winter_tripduration']
data.head()
```

```python
data.plot(x='day', y=['summer_tripduration', 'winter_tripduration'], figsize=(10,10))
plt.title('Tripdurations in summer and winter', fontsize=20)
plt.xlabel('Days of Months', fontsize=15)
plt.ylabel('Average Tripdurations of Each Day', fontsize=15)
plt.ylim(500, 1500)
```

## The graph shows tripdurations in summer are generally higher than tripdurations in winter.
